```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('images/tools.pgm',0)
img2 = cv2.imread('images/swan.pgm',0)
img3 = cv2.imread('images/flower.pgm',0)
histogram1 = cv2.calcHist([img1],[0],None,[256],[0,256])
histogram2 = cv2.calcHist([img2],[0],None,[256],[0,256])
histogram3 = cv2.calcHist([img3],[0],None,[256],[0,256])
plt.subplot(231), plt.plot(histogram1),plt.title('tools')
plt.subplot(232), plt.plot(histogram2),plt.title('swan')
plt.subplot(233), plt.plot(histogram3),plt.title('flower')
plt.show()
```

```python
equ1 = cv2.equalizeHist(img1)
equ2 = cv2.equalizeHist(img2)
equ3 = cv2.equalizeHist(img3)
plt.subplot(231), plt.imshow(equ1,'gray'),plt.title('tools')
plt.xticks([]), plt.yticks([])
plt.subplot(232), plt.imshow(equ2,'gray'),plt.title('swan')
plt.xticks([]), plt.yticks([])
plt.subplot(233), plt.imshow(equ3,'gray'),plt.title('flower')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Otsu's thresholding after Gaussian filtering
# Find the threshold that minimizes the weighted within-class variance. 
blur = cv2.GaussianBlur(img1,(5,5),0)
ret3,th1 = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

# Adaptive thresholding after Gaussian filtering
#adaptive thresholding changes the threshold dynamically
th11 = cv2.adaptiveThreshold(img1,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)

ret,thresh1 = cv2.threshold(th1,200,255,cv2.THRESH_BINARY)#do threshold to do image object
dst = cv2.bitwise_and(thresh1,img1)#multi image object with orginal
ret,thresh2 = cv2.threshold(th1,200,255,cv2.THRESH_BINARY_INV)#do threshold to do image background 
dst1 = cv2.bitwise_and(thresh2, img1)
median = cv2.medianBlur(dst1,5)

adding = cv2.add(dst,median)
plt.subplot(231), plt.imshow(th1),plt.title('otus ')
plt.xticks([]), plt.yticks([])
plt.subplot(232), plt.imshow(adding),plt.title('otus after adding ')
plt.xticks([]), plt.yticks([])
plt.subplot(233), plt.imshow(th11),plt.title('Adaptive')
plt.xticks([]), plt.yticks([])
plt.show()
```
